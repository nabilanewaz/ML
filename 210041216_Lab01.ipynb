Task 5:
Find the mean, median and standard deviation of the 1st column of the given dataset.
Give statistical summary of the dataset as many as possible.

```python
col1 = np.genfromtxt(url, delimiter = ',', usecols = [0], dtype= 'float')

mean_val = np.mean(col1)
median_val = np.median(col1)
std_val = np.std(col1)

print(f"Mean: {mean_val:.2f}, Median: {median_val:.2f}, Std Dev: {std_val:.2f}")

allcol = np.genfromtxt(url, delimiter = ',', usecols = (0,1,2,3), dtype= 'float')

mean_all = np.mean(allcol, axis=0)
median_all = np.median(allcol, axis=0)
std_all = np.std(allcol, axis=0)
min_all = np.min(allcol, axis=0)
max_all = np.max(allcol, axis=0)

print("\nStatistical Summary:")
print(f"Mean:   {mean_all.round(2)}")
print(f"Median: {median_all.round(2)}")
print(f"Std:    {std_all.round(2)}")
print(f"Min:    {min_all.round(2)}")
print(f"Max:    {max_all.round(2)}")
```

Task 3:
Suppose you are an investor and want to invest in housing properties. Use this dataset
to make decision on the location of investing.

```python
grouped = df.groupby(['latitude', 'longitude'])['median_house_value'].mean().reset_index()

top_locations = grouped.sort_values(by='median_house_value', ascending=False).head(10)
print("Top 10 locations to invest:\n", top_locations)
```
